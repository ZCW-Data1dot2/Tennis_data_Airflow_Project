```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
import seaborn as sns 
```

```python
merge_player = pd.read_csv('/Users/suma/dev/airflow_home/Tennis_data_Airflow_Project/Data/WimbledonData.csv',encoding = 'ISO-8859-1')
merge_player['count'] = 0
```

```python
merge_player_w = merge_player[merge_player.winner==True]
merge_player_l = merge_player[merge_player.winner==False]
```

```python
wimb_win = merge_player_w[['tournament','name','count']]
wimb_lose = merge_player_l[['tournament','name','count']]
```

```python
wimb_win
```

```python
wimb_win = wimb_win.groupby(['tournament', 'name']).count()
wimb_win = wimb_win.reset_index()
wimb_win.columns = ['tournament', 'fullname', 'Count_Win']
```

```python
wimb_win
```

```python
wimb_lose = wimb_lose.groupby(['tournament', 'name']).count()
wimb_lose = wimb_lose.reset_index()
wimb_lose.columns = ['tournament', 'fullname', 'Count_Lose']
```

```python
wimb_winner = pd.merge(wimb_win, wimb_lose, on=['tournament', 'fullname'])
```

```python
wimb_winner
```

```python
wimb_winner['total_play'] = wimb_winner['Count_Win'] + wimb_winner['Count_Lose']
wimb_winner['perc_win'] = round(wimb_winner['Count_Win'] / wimb_winner['total_play'],4)*100
wimb_winner = wimb_winner[wimb_winner.total_play > 5]
```

```python
wimb_winner.sort_values(by='perc_win', ascending=False).head(30)
```

```python
top_wimb = wimb_winner[wimb_winner.tournament == 'Wimbledon'].sort_values(by='perc_win', ascending = False).head(10)
```

```python
top_wimb
```

```python
g=sns.catplot(x='fullname', y='perc_win', data=top_wimb, kind='bar', height=5, aspect=2)
g.set_xticklabels(labels = top_wimb.fullname,  rotation=45)
plot.title('Best players in US Open')
plot.savefig("/Users/suma/dev/airflow_home/Tennis_data_Airflow_Project/Images/Top_wimb.png", bbox_inches = 'tight')
```

```python
ax = plot.gca()
top_wimb.plot(x='fullname',y='Count_Win',kind='bar',ax=ax)
top_wimb.plot(x='fullname',y='Count_Lose',kind='bar',color = 'red',ax=ax)
top_wimb_win_lose = top_wimb.copy()

```
