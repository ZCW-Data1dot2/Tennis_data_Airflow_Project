```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
import seaborn as sns 
```

```python
merge_player = pd.read_csv('/Users/suma/dev/airflow_home/Tennis_data_Airflow_Project/Data/USOpenData.csv',encoding = 'ISO-8859-1')
merge_player['count'] = 0
```

```python
merge_player_w = merge_player[merge_player.winner==True]
merge_player_l = merge_player[merge_player.winner==False]
```

```python
USopen_win = merge_player_w[['tournament','name','count']]
USopen_lose = merge_player_l[['tournament','name','count']]
```

```python
USopen_win
```

```python
USopen_win = USopen_win.groupby(['tournament', 'name']).count()
USopen_win = USopen_win.reset_index()
USopen_win.columns = ['tournament', 'fullname', 'Count_Win']
```

```python
USopen_win
```

```python
USopen_lose = USopen_lose.groupby(['tournament', 'name']).count()
USopen_lose = USopen_lose.reset_index()
USopen_lose.columns = ['tournament', 'fullname', 'Count_Lose']
```

```python
USopen_winner = pd.merge(USopen_win, USopen_lose, on=['tournament', 'fullname'])
```

```python
USopen_winner
```

```python
USopen_winner['total_play'] = USopen_winner['Count_Win'] + USopen_winner['Count_Lose']
USopen_winner['perc_win'] = round(USopen_winner['Count_Win'] / USopen_winner['total_play'],4)*100
USopen_winner = USopen_winner[USopen_winner.total_play > 5]
```

```python
USopen_winner.sort_values(by='perc_win', ascending=False).head(30)
```

```python
top_USopen = USopen_winner[USopen_winner.tournament == 'US Open'].sort_values(by='perc_win', ascending = False).head(10)
```

```python
g=sns.catplot(x='fullname', y='perc_win', data=top_USopen, kind='bar', height=5, aspect=2)
g.set_xticklabels(labels = top_USopen.fullname,  rotation=45)
plot.title('Best players in US Open')
plot.savefig("/Users/suma/dev/airflow_home/Tennis_data_Airflow_Project/Images/Top_US.png", bbox_inches = 'tight')
```

```python
ax = plot.gca()
top_USopen.plot(x='fullname',y='Count_Win',kind='bar',ax=ax)
top_USopen.plot(x='fullname',y='Count_Lose',kind='bar',color = 'red',ax=ax)
top_USopen_win_lose = top_USopen.copy()

```
